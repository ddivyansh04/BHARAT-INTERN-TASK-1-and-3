```python
X=df.drop(['date','id','price'],axis=1)
```

```python
X.head()
```

```python
Y=df.price
```

```python
#let's split the data
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(X,Y, test_size = 0.10,random_state =2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
conv_dates = [1 if values == 2014 else 0 for values in df.date ]
df['date'] = conv_dates
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
#as score is not good so let's try other method
```

```python
from sklearn import ensemble

clf = ensemble.GradientBoostingRegressor(n_estimators=400, 
                                         max_depth=5, 
                                         min_samples_split=2,
                                         learning_rate=0.1, 
                                         loss='squared_error')  # Use 'squared_error' instead of 'ls'

# Now you can proceed to train your model
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
#as by this ensemble method we are getting enough score

predictions_reg = reg.predict(x_test)
predictions_clf = clf.predict(x_test)
```

```python
import matplotlib.pyplot as plt

# Assuming y_test, x_test, reg, and clf are defined

# Scatter plot for predictions from reg (Linear Regressor)
plt.scatter(y_test, reg.predict(x_test), color='green', label='Linear Regressor Predictions')

# Scatter plot for predictions from clf (GradientBoostingRegressor)
plt.scatter(y_test, clf.predict(x_test), color='orange', label='GradientBoostingRegressor Predictions')

plt.title('Comparison of Predictions from Linear Regressor and GradientBoostingRegressor')
plt.xlabel('y_test')
plt.ylabel('Predictions')
plt.legend()
plt.grid(True)
plt.show()
```
